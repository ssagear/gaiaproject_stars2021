# All Luminosity Functions

```python

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.optimize as opt
from scipy.optimize import curve_fit

import seaborn as sns

plt.rcParams['figure.figsize'] = [12, 8]
```

```python
M67 = np.genfromtxt("M67lumfunc.csv", delimiter=',')
M44 = np.genfromtxt("M44lumfunc.csv", delimiter=',')
M71 = np.genfromtxt("M71lumfunc.csv", delimiter=',')
NGC188 = np.genfromtxt("NGC188lumfunc.csv", delimiter=',')
NGC6791 = np.genfromtxt("NGC6791lumfunc.csv", delimiter=',')
```

```python
a = 0.8
fsize = 25

plt.hist(M71, density=1, label='M71', color='slateblue', alpha=a)
plt.hist(NGC6791, density=1, label='NGC 6791', color='forestgreen', alpha=a)
plt.hist(M44, density=1, label='M44', color='darkorchid', alpha=a)
plt.hist(NGC188, density=1, label='NGC 188', color='xkcd:bright blue', alpha=a)
plt.hist(M67, density=1, label='M67', color='teal', alpha=a)


plt.xlabel('Absolute V Magnitude', fontsize=fsize)
plt.ylabel('$n_{stars}/n_{stars, total}$', fontsize=fsize)
plt.title('Binned Luminosity Functions', fontsize=fsize)
plt.xticks(fontsize=fsize)
plt.yticks(fontsize=fsize)
plt.legend(fontsize=15)

plt.savefig("LumHist.png")
```

```python
sns.distplot(M71, hist=False, kde=True, 
             bins=25, color = 'slateblue',
             kde_kws={'linewidth': 4},label='M71')
sns.distplot(NGC6791, hist=False, kde=True, 
             bins=25, color = 'forestgreen',
             kde_kws={'linewidth': 4},label='NGC 6791')
sns.distplot(M44, hist=False, kde=True, 
             bins=25, color = 'darkorchid',
             kde_kws={'linewidth': 4},label='M44')
sns.distplot(NGC188, hist=False, kde=True, 
             bins=25, color = 'xkcd:bright blue',
             kde_kws={'linewidth': 4},label='NGC 188')
sns.distplot(M67, hist=False, kde=True, 
             bins=25, color = 'teal',
             kde_kws={'linewidth': 4},label='M67')

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('Absolute V Magnitude', fontsize=25)
plt.ylabel('CDF', fontsize=25)
plt.title('Generalized Luminosity Functions', fontsize=fsize)
plt.legend(fontsize=15)


plt.savefig("LumGenHist.png")
```

```python
kwargs = {'cumulative': True}

sns.distplot(M71, hist_kws=kwargs, kde_kws=kwargs, color = 'slateblue', label='M71')
sns.distplot(NGC6791, hist_kws=kwargs, kde_kws=kwargs, color = 'forestgreen', label='NGC 6791')
sns.distplot(M44, hist_kws=kwargs, kde_kws=kwargs, color = 'darkorchid', label='M44')
sns.distplot(NGC188, hist_kws=kwargs, kde_kws=kwargs, color = 'xkcd:bright blue', label='NGC 188')
sns.distplot(M67, hist_kws=kwargs, kde_kws=kwargs, color = 'teal', label='M67')


plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('Absolute V Magnitude', fontsize=25)
plt.ylabel('CDF', fontsize=25)
plt.title('Luminosity CDFs', fontsize=fsize)
plt.legend(fontsize=15)


plt.savefig("LumCDF.png")
```
