# M71

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.optimize as opt
from scipy.optimize import curve_fit

import seaborn as sns

plt.rcParams['figure.figsize'] = [12, 8]
```

### Define functions

```python
def gminusv(gbp_grp):
    """Given G_BP - G_RP, get G_V.
    From Johnson-Cousins Relationship
    """
    b = -0.01760
    m1 = -0.006860
    m2 = -0.1732
    g_v = m2*(gbp_grp)**2 + m1*(gbp_grp) + b
    return g_v

def gminusI(gbp_grp):
    """Given G_BP - G_RP, get G_I.
    From Johnson-Cousins Relationship
    """
    b = 0.02085
    m1 = 0.7419
    m2 = -0.09631
    g_I = m2*(gbp_grp)**2 + m1*(gbp_grp) + b
    return g_I

def cut_box(length, pmra, pmdec, bounds):
    
    """
    bounds = [ramin, ramax, decmin, decmax]
    
    Returns: None
    """
    for i in range(length):
        if pmra[i] < bounds[0] or pmra[i] > bounds[1] or pmdec[i] < bounds[2] or pmdec[i] > bounds[3]:
            pmra[i] = np.nan
            pmdec[i] = np.nan
            
def gaussian(x, mean, amplitude, standard_deviation):
    return amplitude * np.exp( - (x - mean)**2 / (2*standard_deviation ** 2))


def fit_gauss(x, p0, RA=True, Dec=False, nplot='3'):
    
    bin_heights, bin_borders, _ = plt.hist(x, bins='auto', label='histogram', color='slateblue');
    bin_centers = bin_borders[:-1] + np.diff(bin_borders) / 2
    popt, _ = curve_fit(gaussian, bin_centers, bin_heights, p0=p0)

    x_interval_for_fit = np.linspace(bin_borders[0], bin_borders[-1], 10000)
    plt.plot(x_interval_for_fit, gaussian(x_interval_for_fit, *popt), label='fit', c='deeppink')
    plt.legend()

    plt.xticks(fontsize=20)
    plt.yticks(fontsize=20)
    
    if RA==True:
        plt.xlabel('RA Proper Motions (deg)', fontsize=25)
        plt.title('M71 RA Proper Motions (No Cluster Vetting)', fontsize=25)
        plt.savefig('M71/Plot3_M71_RAProperMotions.png', dpi=150)
        
    elif Dec==True:
        plt.xlabel('Dec Proper Motions (deg)', fontsize=25)
        plt.title('M71 Dec Proper Motions (No Cluster Vetting)', fontsize=25)
        plt.savefig('M71/Plot3_M71_DecProperMotions.png', dpi=150)


    
    return popt


def isInside(circle_x, circle_y, rad, x, y):
        
    """Is x, y inside circle?
    """
    if ((x - circle_x) * (x - circle_x) + 
        (y - circle_y) * (y - circle_y) <= rad * rad):
        return True
    else:
        return False

    
def EV_I(AV):
    """
    Given Av, get E(V-I).
    
    """
    
    return AV/2.5

def V_I0(AV, V_I):
    
    """
    Given Av and V-I, get (V-I) - E(V-I).
    
    """
    
    return V_I-EV_I(AV)

def parallax_dist(mean_parallax):
    """
    Given parallax (arcsec), get distance (parsecs).
    
    """
    
    return 1./mean_parallax

def parallax_dist_errprop(parallax, parallax_err):
    """
    Given parallax error (arcsec), get distance error (parsecs).
    
    """
    return parallax_err/parallax
    

def dist_mag(dist, appmag, Av):
    """
    Distance modulus equation.
    Given distance, apparent mag, and Av, get absolute mag.
    
    """
    
    absmag = -1*(5*np.log10(dist)-5+Av-appmag)
    return absmag


def dist_mag_errprop(dist, delAv, deldist):
    """
    Distance modulus error propogation.
    Given distance, apparent mag, and Av, get absolute mag.
    
    """
    
    delA = 5*0.434*(deldist/dist)
    delM = np.sqrt(delAv**2 + delA**2)
    return delM
```

# Converting Gaia magnitude to V and I bands

```python
# Read CSV of Gaia data
M71df = pd.read_csv('M71-result.csv')
M71df = M71df.dropna()

M71df = M71df.drop(M71df[M71df.parallax < -1].index)
M71df = M71df.drop(M71df[M71df.parallax > 1].index)
```

```python
pmra = np.array(M71df['pmra'])
pmdec = np.array(M71df['pmdec'])

pmra0 = np.array(M71df['pmra'])
pmdec0 = np.array(M71df['pmdec'])

ra0 = np.array(M71df['ra'])
dec0 = np.array(M71df['dec'])
```

```python
# Get G_V and G_I for all stars in sample

g_v = gminusv(M71df['phot_bp_mean_mag']-M71df['phot_rp_mean_mag'])

g_I = gminusI(M71df['phot_bp_mean_mag']-M71df['phot_rp_mean_mag'])
```

```python
# Get V mag and I mag for all stars in sample

V = -1*(g_v - M71df['phot_g_mean_mag'])
I = -1*(g_I - M71df['phot_g_mean_mag'])

# Get V-I mag for all stars in sample

V_I = V-I
```

# Step 4 - Color-Magnitude Diagram

Here, I plot a color-magnitude diagram based on the color (V-I) and V magnitude calculated above.

```python
plt.scatter(V_I, V, s=1.0, c='slateblue')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('V-I (mag)', fontsize=25)
plt.ylabel('V (mag)', fontsize=25)
plt.title('M71 Color-Magnitude Diagram (No Cluster Vetting)', fontsize=25)

plt.savefig('M71/M71_Plot1.png', dpi=150)
```

# Step 6: Determine the Center

```python
init_guess_a = [-36., 200., 10.]

ra_center, _, ra_sig = fit_gauss(pmra, init_guess_a)
```

```python
init_guess_b = [-5., 200., 10.]

dec_center, _, dec_sig = fit_gauss(pmdec, init_guess_b, RA=False, Dec=True)
```

```python
# Convert parallax to arcseconds

M71df.insert(12, "parallax_arcsec", M71df['parallax']/1000., True)
M71df.insert(13, "parallax_err_arcsec", M71df['parallax_error']/1000., True)

M71df.insert(14, "distance_parsec", parallax_dist(M71df['parallax_arcsec']), True)
M71df.insert(15, "distance_err_parsec", parallax_dist_errprop(M71df['parallax_arcsec'], M71df['parallax_err_arcsec']))
```

```python
pmsig = np.mean((abs(ra_sig), abs(dec_sig)))
```

```python
for index, row in M71df.iterrows():
    
    if not isInside(ra_center, dec_center, 3*pmsig, row['pmra'], row['pmdec']):
        M71df = M71df.drop(index)
```

```python
pmra = np.array(M71df['pmra'])
pmdec = np.array(M71df['pmdec'])

g_v = gminusv(M71df['phot_bp_mean_mag']-M71df['phot_rp_mean_mag'])
g_I = gminusI(M71df['phot_bp_mean_mag']-M71df['phot_rp_mean_mag'])

V = -1*(g_v - M71df['phot_g_mean_mag'])
I = -1*(g_I - M71df['phot_g_mean_mag'])

V_I = V-I
```

```python
plt.scatter(V_I, V, s=1.0, c='slateblue')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('V-I (mag)', fontsize=25)
plt.ylabel('V (mag)', fontsize=25)
plt.title('M71 Color-Magnitude Diagram (Post-Cluster Vetting)', fontsize=25)

plt.savefig('M71/M71_Plot4.png', dpi=150)
```

### Define Av

```python
Av = 0.879
```

# Step 9: Stellar Parallax

Get Stellar Distances, Absolute V mag, Absolute I mag, and Absolute V-I for all stars in cluster.

```python
M71_dists = np.zeros(len(M71df))
M71_absV = np.zeros(len(M71df))
M71_absI = np.zeros(len(M71df))
M71_absVI = np.zeros(len(M71df))

M71_disterrs = np.zeros(len(M71df))
M71_absVerrs = np.zeros(len(M71df))
M71_absIerrs = np.zeros(len(M71df))
M71_absVIerrs = np.zeros(len(M71df))

V = np.array(V)
I = np.array(I)

i = 0

for index, row in M71df.iterrows():
    M71_dists[i] = parallax_dist(row['parallax_arcsec'])
    M71_disterrs[i] = parallax_dist_errprop(row['parallax_arcsec'], row['parallax_err_arcsec'])
    
    M71_absV[i] = dist_mag(M71_dists[i], V[i], Av)
    M71_absI[i] = dist_mag(M71_dists[i], I[i], Av)
    M71_absVI[i] = M71_absV[i]-M71_absI[i]
    
    
    i+=1
```

```python
# Get (V-I)_0
V_I_naught = V_I0(Av, V_I)
```

```python
plt.scatter(V_I_naught, M71_absV, s=1, color='slateblue')

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('$(V-I)_0$ (mag)', fontsize=25)
plt.ylabel('Absolute V (mag)', fontsize=25)
plt.title('M71 Color-Magnitude Diagram (Stellar Parallax)', fontsize=25)

plt.savefig('M71/M71_Plot9.png', dpi=150)
```

# Step 10: Cluster Parallax

```python
cluster_dist = 1/parallax_arcs
cluster_dist_err = cluster_dist * (parallax_err_arcs/parallax_arcs)
```

```python
print('Distance to M71: ', cluster_dist, ' parsecs')
print('Distance to M71 Error: ', cluster_dist_err, ' parsecs')
```

```python
M71_absV_c = np.zeros(len(M71df))
M71_absI_c = np.zeros(len(M71df))
M71_absVI_c = np.zeros(len(M71df))

M71_absVerrs_c = np.zeros(len(M71df))
M71_absIerrs_c = np.zeros(len(M71df))
M71_absVIerrs_c = np.zeros(len(M71df))

i = 0

for index, row in M71df.iterrows():
    M71_absV_c[i] = dist_mag(cluster_dist, V[i], Av)
    M71_absI_c[i] = dist_mag(cluster_dist, I[i], Av)
    M71_absVI_c[i] = M71_absV_c[i]-M71_absI_c[i]
    
    i+=1
```

```python
plt.scatter(M71_absVI_c, M71_absV_c, s=1, color='slateblue')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

#lt.xlim(-1, 4)
#lt.ylim(-2, 8)
plt.xlabel('$(V-I)_0$ (mag)', fontsize=25)
plt.ylabel('Absolute V (mag)', fontsize=25)
plt.title('M71 Color-Magnitude Diagram (Cluster Parallax)', fontsize=25)

plt.savefig('M71/M71_Plot10.png', dpi=150)
```
